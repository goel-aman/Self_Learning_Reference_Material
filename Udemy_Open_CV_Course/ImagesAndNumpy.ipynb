```python
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
from PIL import Image  
```

```python
pic = Image.open('download.jpg')
```

```python
pic
```

```python
type(pic)
```

```python
pic_arr = np.asarray(pic)
```

```python
type(pic_arr)
```

```python
pic_arr.shape
```

```python
plt.imshow(pic)
```

```python
# R G B
pic_red = pic_arr.copy()[:,:,0]
print(pic_red)
#RED CHANNEL VALUES 0 - 255 
# Higher Value signifies that portion is going to be white. (white means higher value. so in this case means more red.)
plt.imshow(pic_red,cmap = 'gray')
```

```python
pic_red = pic_arr.copy()[:,:,1]
print(pic_red)
#GREEN CHANNEL VALUES 0 - 255
plt.imshow(pic_red,cmap = 'gray') 
```

```python
pic_red = pic_arr.copy()[:,:,2]
print(pic_red)
#BLUE CHANNEL VALUES 0 - 255
plt.imshow(pic_red,cmap = 'gray') 
```

```python
pic_red = pic_arr.copy()
pic_red[:,:,1] = 0
pic_red[:,:,2] = 0
print(pic_red)
#GREEN CHANNEL VALUES 0 - 255
plt.imshow(pic_red) 
```
